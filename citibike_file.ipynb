```python
import pandas as pd
import numpy as np
import datetime as dt
```

## March 2021 Data

```python
citibike_mar_21_df = pd.read_csv("data/202103-citibike-tripdata.csv", low_memory=False)
pd.set_option('display.max_columns', None)
```

```python
citibike_mar_21_df.sample(10)
```

```python
citibike_mar_21_df["birth year"].value_counts()
```

```python
citibike_mar_21_df["gender"].value_counts()
```

```python
year_mar_21_df=citibike_mar_21_df.loc[(citibike_mar_21_df['birth year'] !=1970) & (citibike_mar_21_df["gender"] !=0)]
year_mar_21_df.head()
```

```python
year_mar_21_df["trip_in_hours"] = year_mar_21_df["tripduration"] / 60
```

```python
pd.to_datetime(year_mar_21_df['starttime'])
pd.to_datetime(year_mar_21_df['stoptime'])
```

```python
year_mar_21_df[year_mar_21_df.isna().any(axis=1)]
```

```python
sample_mar_21_df=year_mar_21_df.sample(frac=0.05, random_state=1).reset_index()
sample_mar_21_df
```

```python
sample_mar_21_df.to_csv("output_data/citibike_nyc_df_mar_21.csv")
```
